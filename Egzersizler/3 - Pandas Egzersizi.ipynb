# The Dataset From This [Address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv)

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
df = pd.read_csv(url, sep = "\t")
df.head()
```

## Kaç ürün $10.00'dan daha pahalıdır?
* Öncelikle değişkeni düzeltmemiz gerekiyor.

```python
df["item_price"][1]
```

```python
floatize = lambda x : float(x[1:-1])

df["item_price"] = df["item_price"].apply(floatize)
df.head()
```

```python
df["item_price"].dtype
```

```python
df[df["item_price"] > 10].nunique()
```

```python
df[df["item_price"] > 10].nunique()["item_name"]
```

## Her ürünün fiyatı nedir? ("item_name" & "item_price" aynı anda yazılsın)

```python
cols = ["item_name","item_price"]

df[cols].head()
```

## "item_name" değişkenine göre nasıl sıralanır?

```python
df.sort_values(by = "item_name").head()
```

## En yüksek fiyata sahip olan ürünün fiyatı nedir?

```python
df.sort_values(by = "item_price", ascending = False).head(1)
```
