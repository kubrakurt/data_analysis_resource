# The Dataset From This [Address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv"

df = pd.read_csv(url, sep = ",")
df.head()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
df.head()
```

## "total_bill" değişkeni için histogram nasıl çizdirilir?

```python
sns.distplot(df["total_bill"]).set(xlabel = "Value", ylabel = "Frequency", title = "Total Bill");
```

## "total_bill" ve "tips" değişkenleri için scatter plot nasıl çizdirilir?

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.pairplot(df);
```

## "day" ve "total_bill" değişkenleri arasındaki ilişki nasıl çizdirilir?

```python
sns.scatterplot(x = "day", y = "total_bill", data = df, hue = "day");
```

## X ekseninde "tip", Y ekseninde "day" ve "sex" değişkeni ile renklendirilmiş grafik nasıl çizdirilir?

```python
sns.stripplot(x = "tip", y = "day", data = df, hue = "sex");
```

```python
sns.boxplot(x = "day", y = "total_bill", data = df, hue = "time");
```
