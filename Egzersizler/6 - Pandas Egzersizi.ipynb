# The Dataset From This [Address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv)

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv"

df = pd.read_csv(url)
df.head()
```

## "Mjob" ve "Fjob" değişkenlerinin baş harfleri büyütülür?

```python
capitalize = lambda x: x.capitalize()

df["Mjob"] = df["Mjob"].apply(capitalize)
df["Fjob"] = df["Fjob"].apply(capitalize)
df.head()
```

## 17 yaşından büyükler alkol kullanabiliyor, gösteriniz.

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False
    
df["legal_drinker"] = df["age"].apply(majority)
df.head()
```

* Lambda fonksiyonu ile de yapılabilir.

```python
majority_lambda = lambda x: x > 17 if True else False

df["legal_drinker_lambda"] = df["age"].apply(majority_lambda)
df.head()
```

## Veri setindeki her sayı 10 ile nasıl çarpılır?

```python
def times10(x):
    if type(x) is int:
        return x*10
    return x

columns = df.columns

for col in columns:
    df[col] = df[col].apply(times10)
    
df.head()
```

```python
df = df.applymap(times10)
df.head()
```
